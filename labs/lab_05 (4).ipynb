### 1. Lea el "total_profit" de todos los meses, muéstrelo usando un gráfico lineal y un gráfico de dispersión.

```python
plt.plot(df['month_number'],df['total_profit'], 'o' )
plt.title=('Grafico de Disperción')
plt.xlabel=('Mes')
plt.show()
```

### 2. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico multilínea.

```python
df[['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']].plot(figsize=(18,5))
plt.title('Ventas', size=18)
plt.show()
```

### 3. Lea los datos de ventas de productos de "facecream" y "facewash" y muéstrelos usando el gráfico de barras.

```python
ejex=[]
ejey_facecream=[]
ejey_facewash=[]
for i in range(1,12+1):
    ejex.append(i)
    ejey_facecream.append(df['facecream'][i])
    ejey_facewash.append(df['facewash'][i])
```

```python
plt.bar(ejex, ejey_facecream)
plt.bar(ejex, ejey_facewash)
plt.title('Ventas de FaceCream y Facewash mes a mes', size=17)
plt.legend(labels=['Ventas de Facecream ', 'Ventas de Facewash'])
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.show()
```

### 4. Lea todos los datos de ventas de productos y muéstrelos utilizando un gráfico box-plot.

```python
import numpy as np
```

```python
fig = plt.figure(figsize =(10, 7))
ax = fig.add_subplot(111)
data_a =df['facecream'] 
data_b =df['facewash'] 
data_c =df['toothpaste'] 
data_d =df['bathingsoap']
data_e=df['moisturizer']
data_f=df['shampoo']

data_2d=[data_a,data_b,data_c,data_d, data_e, data_f]

plt.boxplot(data_2d,labels=['facecream', 'facewash','toothpaste','bathingsoap', 'moisturizer','shampoo'])
plt.title("Ventas de Cada Producto", size=18)
plt.show()
```

### 5. Calcule los datos de ventas totales del año pasado para cada producto y muéstrelos usando un gráfico circular

```python
sumas = [df['facecream'].sum(),df['facewash'].sum(),df['toothpaste'].sum(),df['shampoo'].sum(),df['bathingsoap'].sum(), df['moisturizer'].sum() ]
nombres = ['facecream','facewash','toothpaste','shampoo','bathingsoap', 'moisturizer' ]
plt.pie(sumas, labels=nombres, autopct="%0.1f %%")
plt.axis("equal")
plt.title('Ventas Año Anterior de Cada Producto', size=18)
plt.show()
```
